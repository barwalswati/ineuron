ðŸ’¡ Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2),..., (an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

**Example 1:**
Input: nums = [1,4,3,2]
Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4
So the maximum possible sum is 4

```python
def arrayPairSum(nums):
    nums.sort()  # Sort the array in ascending order
    max_sum = 0
    
    for i in range(0, len(nums), 2):
        max_sum += min(nums[i], nums[i+1])
        
    return max_sum
```

```python
nums = [1, 4, 3, 2]
maximized_sum = arrayPairSum(nums)
print(maximized_sum)
```

ðŸ’¡ Alice has n candies, where the ith candy is of type candyType[i]. Alice noticed that she started to gain weight, so she visited a doctor. 

The doctor advised Alice to only eat n / 2 of the candies she has (n is always even). Alice likes her candies very much, and she wants to eat the maximum number of different types of candies while still following the doctor's advice. 

Given the integer array candyType of length n, return the maximum number of different types of candies she can eat if she only eats n / 2 of them.

Example 1:
Input: candyType = [1,1,2,2,3,3]
Output: 3

Explanation: Alice can only eat 6 / 2 = 3 candies. Since there are only 3 types, she can eat one of each type.

```python
def maxCandies(candyType):
    max_eaten = len(set(candyType))  # Count the number of unique candies
    
    # Alice can only eat n / 2 candies
    max_eaten = min(max_eaten, len(candyType) // 2)
    
    return max_eaten
```

```python
candyType = [1, 1, 2, 2, 3, 3]
max_eaten = maxCandies(candyType)
print(max_eaten) 
```

ðŸ’¡ We define a harmonious array as an array where the difference between its maximum value
and its minimum value is exactly 1.

Given an integer array nums, return the length of its longest harmonious subsequence
among all its possible subsequences.

A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements without changing the order of the remaining elements.

Example 1:
Input: nums = [1,3,2,2,5,2,3,7]
Output: 5

Explanation: The longest harmonious subsequence is [3,2,2,2,3].

```python
def findLHS(nums):
    num_counts = {}
    longest_length = 0
    
    # Count the occurrences of each number
    for num in nums:
        if num in num_counts:
            num_counts[num] += 1
        else:
            num_counts[num] = 1
    
    # Iterate over the dictionary and check for harmonious subsequences
    for num in num_counts:
        if num + 1 in num_counts:
            curr_length = num_counts[num] + num_counts[num + 1]
            longest_length = max(longest_length, curr_length)
    
    return longest_length
```

```python
nums = [1, 3, 2, 2, 5, 2, 3, 7]
longest_subsequence_length = findLHS(nums)
print(longest_subsequence_length)
```

ðŸ’¡ You have a long flowerbed in which some of the plots are planted, and some are not.
However, flowers cannot be planted in adjacent plots.
Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise.

Example 1:
Input: flowerbed = [1,0,0,0,1], n = 1
Output: true

```python
def canPlaceFlowers(flowerbed, n):
    count = 0
    i = 0
    
    while i < len(flowerbed):
        if flowerbed[i] == 0 and (i == 0 or flowerbed[i-1] == 0) and (i == len(flowerbed)-1 or flowerbed[i+1] == 0):
            # Found an empty plot where flowers can be planted
            flowerbed[i] = 1  # Plant a flower
            count += 1
            
        if count >= n:
            return True
        
        i += 1
    
    return False
```

```python
flowerbed = [1, 0, 0, 0, 1]
n = 1
can_plant = canPlaceFlowers(flowerbed, n)
print(can_plant)
```

ðŸ’¡ Given an integer array nums, find three numbers whose product is maximum and return the maximum product.

Example 1:
Input: nums = [1,2,3]
Output: 6

```python
def maximumProduct(nums):
    nums.sort()  # Sort the array in ascending order

    # Calculate the two possible products
    product1 = nums[-1] * nums[-2] * nums[-3]  # Product of three largest numbers
    product2 = nums[0] * nums[1] * nums[-1]  # Product of two smallest negative numbers and the largest number

    # Return the maximum product
    return max(product1, product2)
```

```python
nums = [1, 2, 3]
max_product = maximumProduct(nums)
print(max_product)
```

ðŸ’¡ Given an array of integers nums which is sorted in ascending order, and an integer target,
write a function to search target in nums. If target exists, then return its index. Otherwise,
return -1.

You must write an algorithm with O(log n) runtime complexity.

Input: nums = [-1,0,3,5,9,12], target = 9
Output: 4

Explanation: 9 exists in nums and its index is 4

```python
def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

```python
nums = [-1, 0, 3, 5, 9, 12]
target = 9
index = search(nums, target)
print(index)
```

ðŸ’¡ An array is monotonic if it is either monotone increasing or monotone decreasing.

An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is
monotone decreasing if for all i <= j, nums[i] >= nums[j].

Given an integer array nums, return true if the given array is monotonic, or false otherwise.

Example 1:
Input: nums = [1,2,2,3]
Output: true

```python
def isMonotonic(nums):
    is_increasing = True
    is_decreasing = True

    for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:
            is_increasing = False
        if nums[i] < nums[i + 1]:
            is_decreasing = False

    return is_increasing or is_decreasing
```

```python
nums = [1, 2, 2, 3]
is_monotonic = isMonotonic(nums)
print(is_monotonic)
```

ðŸ’¡ You are given an integer array nums and an integer k.

In one operation, you can choose any index i where 0 <= i < nums.length and change nums[i] to nums[i] + x where x is an integer from the range [-k, k]. You can apply this operation at most once for each index i.

The score of nums is the difference between the maximum and minimum elements in nums.

Return the minimum score of nums after applying the mentioned operation at most once for each index in it.

Example 1:
Input: nums = [1], k = 0
Output: 0

Explanation: The score is max(nums) - min(nums) = 1 - 1 = 0.

```python
def minScore(nums, k):
    min_num = float('inf')
    max_num = float('-inf')

    for num in nums:
        min_num = min(min_num, num)
        max_num = max(max_num, num)

    if max_num - min_num <= 2 * k:
        return 0

    min_score = float('inf')

    for num in nums:
        for x in range(-k, k+1):
            modified_num = num + x
            new_min_num = min(min_num, modified_num)
            new_max_num = max(max_num, modified_num)
            min_score = min(min_score, new_max_num - new_min_num)

    return min_score
```

```python
nums = [1]
k = 0
score = minScore(nums, k)
print(score)
```
