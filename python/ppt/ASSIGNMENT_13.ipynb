**Question 1**

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

**Examples:**

Input: list1 = 5->2->3->8
list2 = 1->7->4->5
Output: New list = 5->7->4->8

Input:list1 = 2->8->9->3
list2 = 5->3->6->4
Output: New list = 5->8->9->4

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def create_new_list(list1, list2):
    # Initialize pointers for list1, list2, and newList
    ptr1 = list1
    ptr2 = list2
    newList = None
    tail = None

    # Traverse both lists
    while ptr1 is not None and ptr2 is not None:
        # Compare values at the current position
        if ptr1.data >= ptr2.data:
            new_node = Node(ptr1.data)
            ptr1 = ptr1.next
        else:
            new_node = Node(ptr2.data)
            ptr2 = ptr2.next

        # Add the greater node to the newList
        if newList is None:
            newList = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    # Add any remaining nodes from list1 or list2
    if ptr1 is not None:
        tail.next = ptr1
    elif ptr2 is not None:
        tail.next = ptr2

    return newList

# Example usage:
# list1: 5->2->3->8
# list2: 1->7->4->5
node1 = Node(5)
node2 = Node(2)
node3 = Node(3)
node4 = Node(8)

node1.next = node2
node2.next = node3
node3.next = node4

node5 = Node(1)
node6 = Node(7)
node7 = Node(4)
node8 = Node(5)

node5.next = node6
node6.next = node7
node7.next = node8

new_list = create_new_list(node1, node5)

# Traverse and print the resulting list
current = new_list
while current is not None:
    print(current.data, end="->")
    current = current.next

# Output: 5->7->4->8
```

Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

**Example 1:**
Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def removeDuplicates(head):
    # Check if the list is empty or has a single node
    if head is None or head.next is None:
        return head

    # Traverse the linked list
    current = head
    while current.next is not None:
        # Compare the current node with the next node
        if current.data == current.next.data:
            # Remove the duplicate node by updating the next pointer
            current.next = current.next.next
        else:
            # Move to the next node
            current = current.next

    return head

# Example usage:
# LinkedList: 11->11->11->21->43->43->60
node1 = Node(11)
node2 = Node(11)
node3 = Node(11)
node4 = Node(21)
node5 = Node(43)
node6 = Node(43)
node7 = Node(60)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7

head = removeDuplicates(node1)

# Traverse and print the resulting linked list
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next

# Output: 11->21->43->60
```

Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**
Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3
Output:   3->2->1->4->5->6->9->8->7->NULL.

```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None


def reverse_alternate_k_nodes(head, k):
    if not head or k <= 1:
        return head

    current = head
    prev = None
    is_reversed = True  

    while current:
        # Skip k nodes
        count = 0
        while current and count < k:
            prev = current
            current = current.next
            count += 1

        # Reverse the alternate segment
        if is_reversed:
            prev.next = reverse_segment(prev.next, k)

        # Toggle the flag for the next segment
        is_reversed = not is_reversed

    return head


def reverse_segment(head, k):
    prev = None
    current = head
    next_node = None
    count = 0

    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed segment to the remaining list
    if next_node:
        head.next = next_node

    return prev

```

Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:
Input:   1->2->3->5->2->10, key = 2
Output:  1->2->3->5->10

```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None


def delete_last_occurrence(head, key):
    if not head:
        return None

    last_occurrence = None
    current = head
    prev = None

    # Traverse the linked list to find the last occurrence of the key
    while current:
        if current.val == key:
            last_occurrence = prev

        prev = current
        current = current.next

    # If the last occurrence is the head, update the head
    if not last_occurrence:
        head = head.next
    else:
        last_occurrence.next = last_occurrence.next.next

    return head
```

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4



```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None


def merge_sorted_lists(a, b):
    # Create a dummy node as the head of the merged list
    dummy = ListNode(0)
    current = dummy

    # Initialize pointers for both lists
    ptr1 = a
    ptr2 = b

    # Merge the lists
    while ptr1 and ptr2:
        if ptr1.val <= ptr2.val:
            current.next = ptr1
            ptr1 = ptr1.next
        else:
            current.next = ptr2
            ptr2 = ptr2.next

        current = current.next

    # Append remaining nodes of the non-empty list
    if ptr1:
        current.next = ptr1
    elif ptr2:
        current.next = ptr2

    # Return the head of the merged list (excluding the dummy node)
    return dummy.next
```

Question 7**

Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**

Original Linked list 10 8 4 2
Reversed Linked list 2 4 8 10

```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None


def reverse_doubly_linked_list(head):
    if not head:
        return None

    current = head

    # Swap the prev and next pointers of each node
    while current:
        temp = current.prev
        current.prev = current.next
        current.next = temp
        current = current.prev

    # Set the head of the original list as the new tail of the reversed list
    if temp:
        head = temp.prev

    return head
```

*Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

**Example 1:Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

example2:Input:
LinkedList = 1 <--> 5 <--> 2 <--> 9
x = 1
Output:5 2 9

```python
class ListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None


def delete_node_at_position(head, position):
    if not head:
        return None

    # Case for deleting the head node
    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    current = head
    count = 1

    # Traverse to the node at the given position
    while current and count < position:
        current = current.next
        count += 1

    # If the given position is out of range
    if not current:
        return head

    # Update the next and previous pointers of the surrounding nodes
    prev_node = current.prev
    next_node = current.next
    if prev_node:
        prev_node.next = next_node
    if next_node:
        next_node.prev = prev_node

    return head
```
